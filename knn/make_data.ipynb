```python
import pandas as pd
import os
import glob

# data_dir='fixed_csv_label_ver2'
data_dir='filtered_csv_label'
current_dir=os.getcwd()
noise_data_dir=os.path.join(current_dir, data_dir)
noise_data_file=os.path.join(noise_data_dir,'-21*.csv')
noise_data_files=glob.glob(noise_data_file)
learning_data_dir=os.path.join(current_dir, data_dir)
noise_data_files
```

```python
file_list=[]
for j in range(len(noise_data_files)):
    df=pd.read_csv(noise_data_files[j], header=None, usecols=[0,1,2],skiprows=4, 
                   names=['original', 'label', 'time'])
    df['day']=j
    file_list.append(df)
df=pd.concat(file_list).reset_index(drop=True)
df
```

```python
df['time']=pd.to_datetime(df['time'], format='%H:%M:%S').dt.hour
df
```

```python
output_file=f'learning_data_10.csv'
df.to_csv(os.path.join(learning_data_dir, output_file), index=False)
```
